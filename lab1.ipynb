```python
# LABORATORIO 1 - INTELIGENCIA ARTIFICIAL

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
#Cargar el data set y visualizar los 5 primeros campos

df = pd.read_csv('census.csv')
df.head()
```

```python
#Tamaño de filas y columnas

nfilas = df.shape[0]
ncolum = df.shape[1]
print ('Número de filas: ', nfilas)
print ('Número de columnas: ', ncolum)
```

```python
# Matriz de correlación 

matriz_c = df.corr()
matriz_c
```

```python
# Limpiar el dataset (no todos los campos son necesarios, solo use los que necesite)

cleandf=['COUNTY', 'CTYNAME']
df_new=df.drop(cleandf, axis =1)

df_new

```

```python
# Codificar la data categorica

import pandas as pd

cat_dat = pd.get_dummies (df_new['STNAME'], prefix='ST')
df_nect = df_new.drop (['STNAME'], axis=1)
df_nect = pd.concat ([df_nect, cat_dat], axis =1)
df_nect
```

```python
X = df_new [['BIRTHS2010', 'CENSUS2010POP']]
Y = df_new [ 'STNAME']
```

```python
from sklearn.model_selection import train_test_split

X_entr, X_prue, Y_entr, Y_prue = train_test_split (X, Y, test_size=0.3, stratify=Y)
X_prue

```

```python

from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
clf.fit (X_prue, Y_prue)
```

```python
Y_pred = clf.predict(X_prue)
Y_pred
```

```python
from sklearn.metrics import accuracy_score, recall_score
accuracy = accuracy_score (Y_prue, Y_pred)
accuracy
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix (Y_prue, Y_pred)
cm
```

```python
from sklearn.neighbors import KNeighborsClassifier

clk = KNeighborsClassifier()

clk.fit(X_entr, Y_entr)
```

```python
y_predkn = clk.predict(X_prue)
y_predkn
```

```python
accuracykn = accuracy_score(Y_prue, y_predkn)
accuracykn
```
